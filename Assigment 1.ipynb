```python
X = X_data['health_training']
Y = X_data['2018_cases_total']

plt.figure(figsize= (10,10))
plt.xlabel('Budget allocated to training health professionals')
plt.ylabel('Counts of those who developed ‘Condition X’ in 2018')

a = plt.plot(X,Y,'.',label ='original data')

slope, intercept, r_value, p_value, stderr = sps.linregress(X,Y)

b = plt.plot(X, X*slope + intercept, 'r', label = 'fitted line')
plt.legend()
plt.title("Fitted regrassion model")
plt.show()

print("y =", round(slope,3), "X +", round(intercept,3))
print("Rsq =", r_value**2)
print("P-value =", p_value)
```
